```python
def average_probs(probs_list, out_name):
    images = os.listdir(probs_list[-1])
    
    for im in images:
        pred = np.zeros((1232, 1624))
        for probs in probs_list:
            #print(im)
            mask = np.load(os.path.join(probs, im))
            #print(mask.min(), mask.max())
            #print(mask.shape)
            pred += np.clip(mask, 0, 1)
        pred /= len(probs_list)
        out = np.array(np.where(pred>0.5, 1, 0)*255, dtype='uint8')
        cv2.imwrite(os.path.join(out_name, im.split(".")[0] + ".png"), out)    
```
